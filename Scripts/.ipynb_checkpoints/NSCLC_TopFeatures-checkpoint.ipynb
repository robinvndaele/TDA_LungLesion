# Topological feature for NSCLS cohort

In this notebook we extract the topological features for the scans in the NSCLS cohort.

```python
# Topological feature extraction
import TDAfeatures as tf

# Our custom functions to conduct our feature comparison experiments
import CV_LIDC
from CV_LIDC import run_exps_reg, run_exps_class

# Handling arrays and data frames
import numpy as np
import pandas as pd 

# Obtaining p-values
import scipy

# Plotting
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading the data

We start by loading the meta data (including the annotated tumor malignancy), the radiomic features, and the topological features. The meta data also includes semantic features that were manually annotated by expert radiologists, which they consecutively used for their malignancy prediction.

### Loading the malignancy scores and  semantic features

We used the pylidc library to obtain the annotation information provided by expert radiologists for all nodules (one patient may have multiple annotated tumor nodules). The resulting data frame can be loaded as follows.

```python
meta = pd.read_csv(os.path.join("Features", "LIDC", "Meta.csv"), index_col=0)
meta.head(5)
```

### Loading the accurate diagnosis data

The malignancy scores assigned by the expert radiologist is a best guess, based on the semantic features above which they manually and visually infer from the scan. Unfortunately, their assigned outcome does not always reflect the true histology of the tumor.

For a small number of patients, LIDC includes more accurate biological diagnoses, which were obtained through one of the following assessments:
- review of radiological images to show two years of stable nodule
- biopsy
- surgical resection
- progression or response

We will use these outcomes to evaluate our features for benign vs. malignant classification, and see how this compares to the radiologist predictions. Similar to the SF/PA cohort, we will only include primary lung tumors. The diagnosis data is loaded as follows.

```python
diagnosis = pd.DataFrame()

diagnosis_temp = pd.read_excel(os.path.join("Features", "LIDC", "Diagnosis.xls"), index_col=0)

for p in diagnosis_temp.index:
    for idx, c in enumerate(diagnosis_temp.columns):
        if diagnosis_temp.loc[p, c] == " " or diagnosis_temp.loc[p, c] != diagnosis_temp.loc[p, c]:
            break
        if c.startswith("Nodule") and not "Method" in c \
            and diagnosis_temp.loc[p, c] not in [0, 3] and diagnosis_temp.loc[p,:][idx + 1] != 0:
            pnod = p + "_nodule" + str(int(c[7]) - 1)
            this_diagnosis = "Benign" if diagnosis_temp.loc[p, c] == 1 else "Malignant"
            diagnosis.loc[pnod, "Diagnosis"] = this_diagnosis
            
del diagnosis_temp

diagnosis.head(5)
```

### Loading the radiomic features

```python
X_rad = pd.read_csv(os.path.join("Features", "LIDC", "Radiomic.csv"), index_col=0, skipfooter=1, engine="python")
X_rad =  X_rad.loc[:,[c for c in X_rad.columns if c.startswith("original")]]
X_rad.index = [pnod.split("_")[0] + "_nodule" + pnod.split("_")[1] for pnod in X_rad.index]
X_rad.head(5)
```

### Extracting the topological features

The topological features are not stored. Rather, we can quickly obtain them through the stored diagrams, which may also be used to potentially explore other machine learning models. We start by loading these diagrams.

```python
dgms = {"img_sub":{}, "img_sup":{}, "img_box_sub":{}, "img_box_sup":{}, "point_cloud":{}}

for patient in os.listdir(os.path.join("Diagram", "LIDC")):
    for nod in os.listdir(os.path.join("Diagram", "LIDC", patient)):
        
        if len(os.listdir(os.path.join("Diagram", "LIDC", patient, nod, "consensus"))) != 15:
            continue
        
        for dgm in os.listdir(os.path.join("Diagram", "LIDC", patient, nod, "consensus")):
        
            dgmtype = "_".join(dgm.split("_")[:-1])
            dgmdim = dgm.split("_")[-1].replace(".npy", "")
            dgms[dgmtype].setdefault(patient + "_" + nod, {})
            dgms[dgmtype][patient + "_" + nod][dgmdim] = \
                np.load(os.path.join("Diagram", "LIDC", patient, nod, "consensus", dgm))
```

We can now obtain our data of topological feature vectors as follows.

```python
X_top = dict()

for dgm_type in dgms.keys():
    for pnod in dgms[dgm_type].keys():
        
        X_top[pnod] = X_top.setdefault(pnod, {})
        
        for dim in dgms[dgm_type][pnod].keys():
            
            features = tf.persistence_statistics(dgms[dgm_type][pnod][dim])
            for key in features.keys():
                new_key = dgm_type + "_" + dim + "_" + key
                X_top[pnod][new_key] = features[key]
            
            del features

X_top = pd.DataFrame(X_top).transpose()            
X_top.head(5)
```

Some columns of our topological features are known to be constant in advance. For example, the filtration constructed from the image with boundary pixels will always end at one connected component with no higher-dimensional holes. We will discard these from our topological features. Notice that indeed only columns from which we know in advance they are constant are discarded below. 

```python
constant_features = []

for col in X_top.columns:
    values = np.unique(X_top[col])
    values = values[~np.isnan(values)]
    if len(values) == 1:
        constant_features.append(col)
        
X_top = X_top.drop(columns=constant_features)

print("Discarded features (with constant values):")
print("\n")
for f in constant_features:
    print(f)
```

### Final preparations

Finally, we ensure that the indexing of all dataframes is consistent.

```python
missing_rad = np.setdiff1d(meta.index, X_rad.index)
missing_top = np.setdiff1d(meta.index, X_top.index)
patients = np.setdiff1d(meta.index, np.concatenate([missing_rad, missing_top]))
diagnosis = diagnosis.loc[np.intersect1d(diagnosis.index, patients),:]

print("Number of nodules: " + str(len(patients)) +
      " (" + str(len(meta.index) - len(patients)) + " omitted from study)")
print("Accurate diagnosis for " + str(len(diagnosis.index)) + " nodules")
print("Number of radiomic features: " + str(X_rad.shape[1]) + 
      " (missing rows for " + str(len(missing_rad)) + " nodules)")
print("Number of topological features: " + str(X_top.shape[1]) 
      + " (missing rows for " + str(len(missing_top)) + " nodules)")

meta = meta.loc[patients,:]
X_rad =  X_rad.loc[patients,:]
X_top = X_top.loc[patients,:]
```

## Analysis of our outcome by contrast

Below we analyse the distribution of our continuous malignancy scores in our data. This will be used as our dependent variable. We will also separate our analyses for scans obtained with and without contrast material injected into the bloodstream.

```python
with_contrast = [p for p in meta.index if meta.loc[p, "Contrast"] == "Y"]
without_contrast = [p for p in meta.index if meta.loc[p, "Contrast"] == "N"]

print("Number of nodules from scans with contrast: " + str(len(with_contrast)))
print("Number of nodules from scans without contrast: " + str(len(without_contrast)))

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
ax = axes.flatten()

sns.histplot(data=meta.loc[with_contrast,:], x="malignancy", ax=ax[0], bins=10)
sns.histplot(data=meta.loc[without_contrast,:], x="malignancy", ax=ax[1], bins=10)

fig.tight_layout()
plt.show()
```

We also investigate the distribution of our available accurate biologal diagnoses by contrast.

```python
diagnosis_with = np.intersect1d(with_contrast, diagnosis.index)
diagnosis_without = np.intersect1d(without_contrast, diagnosis.index)

all_classes = {"benign": dict(), "malignant": dict(), "Total": dict()}
all_classes["benign"]["Y"] = np.sum(diagnosis.loc[diagnosis_with, "Diagnosis"] == "Benign")
all_classes["benign"]["N"] = np.sum(diagnosis.loc[diagnosis_without, "Diagnosis"] == "Benign")
all_classes["malignant"]["Y"] = np.sum(diagnosis.loc[diagnosis_with, "Diagnosis"] != "Benign")
all_classes["malignant"]["N"] = np.sum(diagnosis.loc[diagnosis_without, "Diagnosis"] != "Benign")
all_classes["Total"]["Y"] = len(diagnosis_with)
all_classes["Total"]["N"] = len(diagnosis_without)
all_classes = pd.DataFrame(all_classes).transpose()
all_classes["Total"] = all_classes["Y"] + all_classes["N"]
    
def grey(s):
    color = "lightgrey"
    return "background-color: %s" % color
    
all_classes.style.applymap(grey, subset=["Total"])
```

## Evaluating features for tumor type prediction

We now compare the effectivess of using radiomic features, topological features, and radiomic + topological  features (through concatenation or ensemble models) for predicting the (continuous) malignancy level or the (binary) more accurate diagnosis. All scores will be evaluathed through the r2 coefficient of determination (regression) or ROC AUC (classification). We will also include the semantic features for comparison. Note that these are the manually annotated features on which the expert radiologist base their malignancy prediction. As such, these will not be highlighted when reaching optimal performance. 

A dictionary will be used to distinguish between categories of features that should be evaluated.

```python
X_dict = {"rad": X_rad, "top": X_top}
```

### Comparing features for malignancy prediction (regression + all nodules)

We compare the effectivess of radiomic features, topological features, and radiomic + topological features, for predicting the continuous malignancy outcome assigned by the expert radiologists.

#### Images with contrast

```python
print("Calculating performances for malignancy prediction\n")

results, regressors, results_summary = run_exps_reg(X_dict={key:X_dict[key].loc[with_contrast,] 
                                                            for key in X_dict.keys()},
                                                    X_base=meta.loc[with_contrast,:]\
                                                                .drop(["malignancy", "Contrast"], axis=1),
                                                    y=meta.loc[with_contrast, "malignancy"], random_state=42)

p = scipy.stats.ttest_ind(results.loc[np.where(results["type"] == "rad")[0], "score"], 
                          results.loc[np.where(results["type"] == "vote_soft")[0], "score"])[1]
if results_summary.loc["mean", "vote_soft score"] > results_summary.loc["mean", "rad score"]:
    p /= 2
else:
    p = 1 - p/2

print("\np-value for difference in performance (rad >= vote): " + str(p))

results_summary.style.highlight_max(color="lightgreen", axis=1, subset=[c for c in results_summary 
                                                                        if c.endswith("score") 
                                                                        and not c.startswith("base")])\
    .applymap(grey, subset=[c for c in results_summary.columns if c.endswith("std")])
```

#### Images without contrast

```python
print("Calculating performances for malignancy prediction\n")

results, regressors, results_summary = run_exps_reg(X_dict={key:X_dict[key].loc[without_contrast,] 
                                                            for key in X_dict.keys()},
                                                    X_base=meta.loc[without_contrast,:]\
                                                            .drop(["malignancy", "Contrast"], axis=1),
                                                    y=meta.loc[without_contrast, "malignancy"], random_state=42)

p = scipy.stats.ttest_ind(results.loc[np.where(results["type"] == "rad")[0], "score"], 
                          results.loc[np.where(results["type"] == "vote_soft")[0], "score"])[1]
if results_summary.loc["mean", "vote_soft score"] > results_summary.loc["mean", "rad score"]:
    p /= 2
else:
    p = 1 - p/2

print("\np-value for difference in performance (rad >= vote): " + str(p))

results_summary.style.highlight_max(color="lightgreen", axis=1, subset=[c for c in results_summary 
                                                                        if c.endswith("score") 
                                                                        and not c.startswith("base")])\
    .applymap(grey, subset=[c for c in results_summary.columns if c.endswith("std")])
```

### Comparing features for malignancy prediction (classification + nodules  with diagnosis only)

We compare the effectivess of radiomic features, topological features, and radiomic + topological features, for predicting the accurate diagnosis. The folds for cross-validation will again be obtained on a patient level. However, the diagnoses are assigned on a nodule level. To ensure there are sufficient examples of each class in each fold, we will perform some stratification over the patients, where each patient is assigned to and indicator whether he or she has a malignant tumor nodule. (Note that in the diagnosis data, a patient may rarely have a different diagnosis for different nodules).

#### Images with contrast

```python
patient_diagnosis = dict()

for pnod in diagnosis_with:
    p = pnod.split("_")[0]
    patient_diagnosis[p] = patient_diagnosis.setdefault(p, [0, 0])
    if diagnosis.loc[pnod, "Diagnosis"] == "Benign":
        patient_diagnosis[p][0] += 1
    else:
        patient_diagnosis[p][1] += 1
    
print("Maximum number of distinct nodule diagnoses per patient: " 
      + str(np.max([np.sum([c != 0 for c in patient_diagnosis[p]]) for p in patient_diagnosis.keys()])))

y_diagnosis_with = [0 if d == "Benign" else 1 for d in diagnosis.loc[diagnosis_with, "Diagnosis"]]
stratify_with = [0 if patient_diagnosis[p][0] == 0 else 1 for p in patient_diagnosis.keys()]
```

```python
print("Calculating performances for malignancy classification\n")

results, classifiers, results_summary = run_exps_class(X_dict={key:X_dict[key].loc[diagnosis_with,] 
                                                               for key in X_dict.keys()},
                                                       X_base=meta.loc[diagnosis_with,:]\
                                                               .drop(["malignancy", "Contrast"], axis=1),
                                                       y=y_diagnosis_with, stratify=stratify_with, 
                                                       random_state=42)

p = scipy.stats.ttest_ind(results.loc[np.where(results["type"] == "rad")[0], "score"], 
                          results.loc[np.where(results["type"] == "vote_soft")[0], "score"])[1]
if results_summary.loc["mean", "vote_soft score"] > results_summary.loc["mean", "rad score"]:
    p /= 2
else:
    p = 1 - p/2

print("\np-value for difference in performance (rad >= vote): " + str(p))

results_summary.style.highlight_max(color="lightgreen", axis=1, subset=[c for c in results_summary 
                                                                        if c.endswith("score") 
                                                                        and not c.startswith("base")])\
    .applymap(grey, subset=[c for c in results_summary.columns if c.endswith("std")])
```

#### Images without contrast

```python
patient_diagnosis = dict()

for pnod in diagnosis_without:
    p = pnod.split("_")[0]
    patient_diagnosis[p] = patient_diagnosis.setdefault(p, [0, 0])
    if diagnosis.loc[pnod, "Diagnosis"] == "Benign":
        patient_diagnosis[p][0] += 1
    else:
        patient_diagnosis[p][1] += 1
    
print("Maximum number of distinct nodule diagnoses per patient: " 
      + str(np.max([np.sum([c != 0 for c in patient_diagnosis[p]]) for p in patient_diagnosis.keys()])))

y_diagnosis_without = [0 if d == "Benign" else 1 for d in diagnosis.loc[diagnosis_without, "Diagnosis"]]
stratify_without = [0 if patient_diagnosis[p][0] == 0 else 1 for p in patient_diagnosis.keys()]
```

```python
print("Calculating performances for malignancy classification\n")

results, classifiers, results_summary = run_exps_class(X_dict={key:X_dict[key].loc[diagnosis_without,] 
                                                               for key in X_dict.keys()},
                                                       X_base=meta.loc[diagnosis_without,:]\
                                                                   .drop(["malignancy", "Contrast"], axis=1),
                                                       y=y_diagnosis_without, stratify=stratify_without, 
                                                       random_state=42)

p = scipy.stats.ttest_ind(results.loc[np.where(results["type"] == "rad")[0], "score"], 
                          results.loc[np.where(results["type"] == "vote_soft")[0], "score"])[1]
if results_summary.loc["mean", "vote_soft score"] > results_summary.loc["mean", "rad score"]:
    p /= 2
else:
    p = 1 - p/2

print("\np-value for difference in performance (rad >= vote): " + str(p))

results_summary.style.highlight_max(color="lightgreen", axis=1, subset=[c for c in results_summary 
                                                                        if c.endswith("score") 
                                                                        and not c.startswith("base")])\
    .applymap(grey, subset=[c for c in results_summary.columns if c.endswith("std")])
```
